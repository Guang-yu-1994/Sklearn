AR自回归模型 
1.   是用自身的数据来预测
2.   必须具有平稳性
3.   必须具有自相关性，如果自相关系数小于0.5，则不宜采用
4.   仅仅适用于预测与自身前期相关的现象


  
  
  

MA移动平均模型
1. 关注自回归模型中的误差项累加
2. 能有效消除预测中的随机波动

I 就是差分项

```
import sys
import os
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.tsa.api as smt
```

```
pd.set_option('display.float.format', lambda x:'%.5f'%x)
np.set_printoptions(precision=5,suppress=True)
pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
```

```
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(stock[asset],lags=20,ax=ax1)
ax1.xaxis.set_ticks_position('bottom')
fig.tight_layout()

ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(stock[asset],lags=20,ax=ax2)
ax2.xaxis.set_ticks_position('bottom')
fig.tight_layout()
```

1. AR(p)看pacf 在第1阶落入置信区间,p取1
2. MA(q)看acf 在第1阶落入置信区间,q取1
3. 差分法确定d


```
train_set = stock[asset].loc[:'2019-11-01',:]
test_set = stock[asset].loc['2019-11-01':,:]
model = sm.tsa.SARIMAX(train_set,order=(2,0,0)).fit()
```

```
import itertools

p_min = 0
d_min = 0
q_min = 0
p_max = 4
d_max = 0
q_max = 4

results_bic = pd.DataFrame(index=['AR{}'.format(i) for i in range(p_min,p_max+1)],
               columns=['MA{}'.format(i) for i in range(q_min,q_max+1)])

for p,d,q in itertools.product(range(p_min,p_max+1),
                 range(d_min,d_max+1),
                 range(q_min,q_max+1)):             
  if p==0 and d==0 and q==0:
    results_bic.loc['AR{}'.format(p),'MA{}'.format(q)] = np.nan 
    continue
  
  try:
    model = sm.tsa.SARIMAX(train_set, order=(p,d,q))
    results = model.fit()
    results_bic.loc['AR{}'.format(p),'MA{}'.format(q)] = results.bic 
  except:
    continue
results_bic = results_bic[results_bic.columns].astype(float)
```

```
fig, ax = plt.subplots(figsize=(10,8))
ax = sns.heatmap(results_bic,
          mask=results_bic.isnull(),
          ax=ax,
          annot=True,
          fmt='.2f')
ax.set_title('BIC')
```

BIC Bayesian Information Criterion  BIC = kln(n) - 2ln(L)， BIC越小模型越好
  K为模型参数个数，n为样本数量,L为似然函数

```
train_results = sm.tsa.arma_order_select_ic(train_set, ic=['aic','bic'],trend='nc',max_ar=4,max_ma=4)

print('AIC', train_results.aic_min_order)
print('BIC', train_results.bic_min_order)
```
