```
data = pd.read_csv('Weather.csv',parse_dates=['date'],index_col='date')
```

```
data.head()
```

```
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```
sns.distplot(data['max_wind_speed'])
```

```
sns.distplot(data['max_wind_speed'],kde=False,rug=True)
```

```
sns.distplot(data['max_wind_speed'],hist=False,rug=True)
```

```
sns.distplot(data['max_wind_speed'],kde=True)
```

```
sns.kdeplot(data['max_wind_speed'],cumulative=True)
```

```
ax = sns.distplot(data['max_wind_speed'])

# Highlight mean
ax.axvline(data['max_wind_speed'].mean(), color='b')

# Highlight median
ax.axvline(data['max_wind_speed'].median(), color='g')
```

```
case = pd.read_csv('Case.csv')
```

```
case.head()
```

```
plt.figure(figsize=(18,10))
plt.xticks(rotation=45,ha='right')
sns.countplot(x='province',hue='group',data=case)
```

```
plt.figure(figsize=(18,10))
plt.xticks(rotation=45,ha='right')
sns.barplot(x='province',y='confirmed',data=case)
```

```
plt.figure(figsize=(18,10))
plt.xticks(rotation=45,ha='right')
sns.pointplot(x='province',y='confirmed',hue='group',data=case)
```

```
plt.figure(figsize=(18,10))
plt.xticks(rotation=45,ha='right',fontsize=12)
plt.ylim([0,1000])
sns.boxplot(x='province',y='confirmed',data=case,hue='group')
```

```
plt.figure(figsize=(18,10))
plt.xticks(rotation=45,ha='right')
plt.ylim([0,1000])
sns.swarmplot(x='province',y='confirmed',data=case)
```
