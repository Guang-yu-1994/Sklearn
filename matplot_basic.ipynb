```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
x = np.linspace(0,100,1000)
y = np.sin(x)
```

```
plt.plot(x,y,color='r',lw=2,ls='-') # ls line style; lw line width
```

```
plt.plot(x,y,c='r',marker='o',markersize=10,markeredgecolor='blue',markerfacecolor='black',label='y=sin(x)',ls='--')
```

```
x = np.arange(0,1.1,0.1)
y = x ** 2
```

```
plt.figure(figsize=(6.4,4.84),dpi=100,facecolor='blue') # dpi分辨率,facecolor边框色
plt.title('y=sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xticks([0,0.2,0.4,0.6,0.8,1])
plt.yticks([0,0.2,0.4,0.6,0.8,1])
plt.plot(x,y,label='y=x**2')
plt.legend(loc='best')
plt.savefig(r'C:\Users\WMX\Desktop\CV\sinx.png')
plt.show()
```

```
data = pd.read_excel('Sales Recorder - access version.xlsm',sheet_name='ProductInfo')
data.columns = ['ID','pID','type','ProName','Size','Price']
```

```
data.head()
```

```
# autopct数据百分比格式
# pctdistance百分数到圆心距离
# wedgeprops={'linewidth':1.5,'edgecolor':'green'} linewidth每块的间距，edgecolor间距填充颜色
# pad 标题距离坐标轴的距离
# explode 突出显示
explode = [0,0.1,0,0.2,0]
plt.pie(x=data.type.value_counts(),explode=explode,labels=data.type.value_counts().keys(),autopct='.1f%%',pctdistance=0.5,\
        labeldistance=1.1,radius=1.5,counterclock=False,startangle=120,\
        wedgeprops={'linewidth':1.5,'edgecolor':'green'},\
        textprops={'fontsize':10,'color':'black'}) 
plt.title('type',pad=30)
```

```
x = data.type.value_counts().keys()
y = data.type.value_counts().values
plt.bar(x=x,height=y,align='center',color='y',tick_label=y)
plt.xlabel('type',labelpad=20)
plt.ylabel('count',labelpad=20)
plt.title('types',pad=30)
```

```
import pandas_datareader.data as web
from datetime import datetime
start = datetime(2018,1,1)
end = datetime(2020,6,30)
asset_list = ['AAPL','GM','TSLA']
stocks = web.DataReader(asset_list,'yahoo',start,end)['Close']
```

```
stocks.head()
```

```
#　density 是否以频数的形式展示
plt.hist(stocks['AAPL'],bins=35,edgecolor='black',density=True)
```

```
plt.scatter(x=stocks['AAPL'],y=stocks['TSLA'],marker='+')
plt.xlabel('AAPL',fontsize=12)
plt.ylabel('TSLA',fontsize=12)
```

```
plt.plot(stocks['TSLA'],label='TSLA')
plt.legend(loc='best')
fig=plt.gcf()
plt.grid(ls='--',c='darkblue') # 虚线网格
fig.set_size_inches(18.5,10.5)
plt.xticks(rotation=45,ha='right')
plt.axhline(y=800,c='r',lw=2) # 平行x轴的参考线
plt.axvline(x=datetime(2020,1,1),c='r',lw=2) # 平行y轴参考线
plt.axvspan(xmin=datetime(2019,1,1),xmax=datetime(2020,1,1),facecolor='green') #　绘制x参考区域
plt.axhspan(ymin=200,ymax=700,facecolor='yellow') #　绘制y参考区域
plt.show()
```

```
fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(111)
ax1.plot(stocks['AAPL'],label='AAPL')
ax1.set_ylabel('stockPrice')
plt.legend(loc='best')
ax2 = ax1.twinx()
ax2.plot(stocks['GM'],label='GM',color='r')
plt.legend(loc='upper center')
plt.show()
```
